```python
import cv2

import numpy as np

import os

import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.patches as patches

import matplotlib.image as mpimg


from matplotlib import cm

import preprocessing

import importlib
```

```python
from itertools import starmap

import numpy as np

# import seaborn as sns

from pathlib import Path

import matplotlib.pyplot as plt
%matplotlib inline

```

```python
folder = 'cfps'

img_names = [i for i in os.listdir(folder)]

```

```python


for img_name in img_names:

        
    img_path = os.path.join(folder, img_name)

    img = cv2.imread(img_path)

    
    img_rgb = img.copy()
        
    img_gray = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY)
    
    xy_size = 0.1
    
    img_gray = cv2.resize(img_gray, (0,0), fx=xy_size, fy=xy_size)
    
    
    H, W = img_gray.shape
    
    _, _, inner_bb_adj, _, outer_bb_adj = preprocessing.get_shapes(img_gray)



    # Create figure and axes
    fig, ax = plt.subplots()

    # Display the image
    ax.imshow(img_gray, cmap='gray')#
    
    ax.axis('off')

    # Create a Rectangle patch
    
    inner_x = int(inner_bb_adj[0] * W)
    
    inner_y = int(inner_bb_adj[1] * H)
    
    inner_w = int(inner_bb_adj[2] * W)
    
    inner_h = int(inner_bb_adj[3] * H)
    
    
    inner_rect = patches.Rectangle((inner_x, inner_y), inner_w, inner_h, linewidth=3, edgecolor='r', facecolor='none')

    # Add the patch to the Axes
    
    ax.add_patch(inner_rect)
   
    outer_x = int(outer_bb_adj[0] * W)
    
    outer_y = int(outer_bb_adj[1] * H)
    
    outer_w = int(outer_bb_adj[2] * W)
    
    outer_h = int(outer_bb_adj[3] * H)
    
    outer_rect = patches.Rectangle((outer_x, outer_y), outer_w, outer_h, linewidth=3, edgecolor='g', facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(outer_rect)
    
    
```
